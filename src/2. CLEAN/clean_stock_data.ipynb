```python
import polars as pl
```

```python
raw_data = pl.read_csv("../../data/raw/raw_stock_data.csv")
raw_data.head(2)
```

```python
clean_data = (
    raw_data
        .rename(
            {"Date": "date",
             "Open": "open",
             "High": "high",
             "Low": "low",
            "Close": "close",
            "Volume": "volume"
            }
        )
        .with_columns(pl.col("date").str.strptime(pl.Datetime))
)
clean_data.write_parquet("../../data/clean/clean_stock_data.parquet")
clean_data.head(2)
```

```python
clean_data.plot("date", "close")
```
